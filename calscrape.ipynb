```python
import requests
```

```python
req = requests.get("https://www.ccny.cuny.edu/registrar/fall-2018-academic-calendar")
```

```python
from bs4 import BeautifulSoup
```

```python
soup = BeautifulSoup(req.text, 'html.parser')
```

```python
soup.title
```

```python
colhead = ['dates', 'date of week', 'text']
```

```python
all_tables = soup.find_all('table')
```

```python
from IPython.display import HTML
```

```python
HTML(str(all_tables[0]))
```

```python
rows = [row for row in all_tables[0].find_all('tr')]
rows[1].find_all('td')
```

```python
value_rows = [[col.get_text().strip().replace('\n', ' ') for col in row.find_all('td')] for row in rows[1:]]
```

```python
value_rows_tidy = []
for value_row in value_rows:
    if value_row[-1] != '':
        value_rows_tidy.append(value_row)
```

```python
value_series =list(zip(*value_rows_tidy))
```

```python
data_dict = dict(zip(colhead, value_series))
```

```python
data_dict
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(data_dict)
```
